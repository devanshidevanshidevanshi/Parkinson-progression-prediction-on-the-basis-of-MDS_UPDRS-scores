```python
'''
General conclusion:

Of the 4 datasets:
The 2 data sets related to the patient's protein content, proteins and peptides, do not contain null values
2 data sets related to the UPDRS score and the patient's medication status, clinical and supplemental, contain null values.
Clinical data set:
There are 51.8% of data lines containing null values ​​-> huge amount of missing data
The missing data is mainly concentrated in 2 attributes updrs_4 and upd23b_clinical_state_on_medication
The two attributes pair updrs_4 and upd23b_clinical_state_on_medication have a very high missing correlation (0.8).
Supplemental dataset:
There are 63.2% of data lines containing null values ​​-> huge amount of missing data
The missing data is mainly concentrated in 2 attributes updrs_4 and upd23b_clinical_state_on_medication, followed by 2 attributes updrs_1 and updrs_2.
The two attributes updrs_4 and upd23b_clinical_state_on_medication have a missing correlation that is not as high as in the clinical data set (0.3).
The pair of 2 attributes updrs_1 and updrs_2 have a very high missing correlation (1.0) -> Whenever updrs_1 is missing, updrs_2 will be missing


Analyze and suggest solutions:

For UPDRS data, specifically updrs_1, updrs_2, updrs_3 and updrs_4:
These columns are the scores that evaluate the patient by symptoms in the categories of the UPDRS, including:
updrs_1: mood and behavior assessment
updrs_2: evaluate daily activities
updrrs_3: assessment of motor functions
updrs_4: assessment of complications of treatment
Important: Higher rating scores indicate correspondingly more severe symptoms. And a score of 0 means that the patient is normal (a value of 0 is significant). Therefore, it is not possible to treat null as a value of 0.
Causes of missing data (hypothesis):
Data collection failed to collect/forgot to collect UPDRS assessment test scores in some patients.
Patient did not undergo UPDRS assessment -> no data to collect.
Patient refuses to provide UPDRS assessment information
Solution directions:
Remove rows containing null values ​​in the UPDRS evaluation columns, or
Assign the null value in the UPDRS evaluation columns to a specific value, provided that when we look at that value, we can know that this value was once null, for example, the value -1.
For the data of the upd23b_clinical_state_on_medication column:
This column records whether the patient is taking Levodopa during the evaluation and receives only 1 of 2 values ​​of On or Off. A null value appearing in this column does not indicate whether the patient is taking the drug or not.
Causes of missing data (hypothesis):
The data collection process failed to collect/forgot to collect the medication status in some patients.
The patient refused to provide medication information.
Solution directions:
Remove column upd23b_clinical_state_on_medication, or
Remove rows with null values ​​in column upd23b_clinical_state_on_medication, or
Assign the null value in the upd23b_clinical_state_on_medication column with the value Unknown, which means that the patient's drug status is unknown. Thus, the column upd23b_clinical_state_on_medication will have 3 values: On, Off and Unknown.
'''
```
